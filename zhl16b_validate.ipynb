# Validate against GFDECO.F by Eric Baker

```python
# Initialize decompression context.
ctx = deco.DecoZHL16B()
                
# Simulate dive profile.
ctx.update_asc_desc(
    time = 5,
    p_start = ctx.p_surface,
    p_end = ctx.depth_to_pressure(50),
    f_n2 = 0.44,
    f_he = 0.35)

ctx.update_constant(
    time = 20,
    p_ambient = ctx.depth_to_pressure(50),
    f_n2 = 0.44,
    f_he = 0.35)

# Get deco info.
ctx.decotime_update(
    p_ambient = ctx.depth_to_pressure(50),
    f_n2 = 0.44,
    f_he = 0.35)

assert ctx.time_to_surface == 57
assert ctx.stops[0]['depth'] == 33
assert ctx.stops[0]['time'] == 1
assert ctx.stops[1]['depth'] == 30
assert ctx.stops[1]['time'] == 1
assert ctx.stops[2]['depth'] == 27
assert ctx.stops[2]['time'] == 1
assert ctx.stops[3]['depth'] == 24
assert ctx.stops[3]['time'] == 1
assert ctx.stops[4]['depth'] == 21
assert ctx.stops[4]['time'] == 1
assert ctx.stops[5]['depth'] == 18
assert ctx.stops[5]['time'] == 1
assert ctx.stops[6]['depth'] == 15
assert ctx.stops[6]['time'] == 2
assert ctx.stops[7]['depth'] == 12
assert ctx.stops[7]['time'] == 4
assert ctx.stops[8]['depth'] == 9
assert ctx.stops[8]['time'] == 7
assert ctx.stops[9]['depth'] == 6
assert ctx.stops[9]['time'] == 12
assert ctx.stops[10]['depth'] == 3
assert ctx.stops[10]['time'] == 24
```
